## 5. Группировка данных, агрегатные функции

**Группировка 1 (среднее)**

```python
year_reiting = df.groupby('year').agg(mean_imdb_score=('imdb_score', 'mean')).sort_values(by = 'mean_imdb_score', ascending = False)
year_reiting
```

Выше представлена таблица, в которой каждому году соответствует среднее значением рейтинга IMDb фильмов, вышедших в этом году, отсортированная от года с наивысшим средним значением к наименьшему.

**Вывод:** Средний рейтинг фильмов вышедших, в **2014 - 2017 г.** выше рейтинга более современных фильмов **2018 - 2021 г.**, поэтому имеет смысл обратить внимание на более ранние картины, получившие высокие оценки зрителей, и менее популярные современные фильмы, чтобы сформировать представление о предпочтениях пользователей.

**Группировка 2 (максимум)**

```python
year_max_reiting = df.groupby('year').agg(max_imdb_score=('imdb_score', 'max')).sort_values(by = 'max_imdb_score', ascending = False)
year_max_reiting
```

В данной таблице каждому году соответствует максимальный рейтинг фильма, вышедшего в том же году.

Мы можем сделать вывод о том, что фильмы набравшие самые высокие оценки были выпущены в **2020, 2018 и 2015 г.** Поэтому, несмотря на высокие средние оценки более ранних фильмов, современные картины удовлетворяют запросам пользователей, благодаря чему мы можем ориентироваться на текущие тенденции в кинематографе при формировании контентной стратегии.

**Группировка 3 (сумма)**

**Диаграмма 2. "Количественное соотношение жанров"**

```python
count_genre = df.groupby('genre').size().reset_index(name='counts').sort_values(by='counts', ascending=False)
```

```python
total_films = 583 #Количество всех фильмов
other_films = 0.015 * total_films # 1.5% от кол-ва всех фильмов
other_genres = count_genre[count_genre['counts'] < other_films]['genre'] #Получение части таблицы со значениями, которые составляют менее 1.5% от общего кол-ва фильмов
other = count_genre[count_genre['genre'].isin(other_genres)]['counts'].sum() #Суммарное количество фильмов из other_genres
```

```python
other_row = pd.DataFrame({'genre': ['other_genres'], 'counts': [other]}) #Строка с информацией об общем кол-ве фильмов с жанрами, составляющих менее 1.5% от общего кол-ва
genres = count_genre[~count_genre['genre'].isin(other_genres)] #Получение части таблицы со значениями, составляющими более 1.5% от общего кол-ва
count_genre = pd.concat([genres, other_row], ignore_index=True) #Объединение строки other_row и таблицы genres
```

```python
#Создание диаграммы
plt.figure(figsize = (6, 6))
count_genre.plot(kind = 'pie', y = 'counts', labels = count_genre['genre'].values, autopct = '%1.1f%%', startangle = 180)
plt.legend().set_visible(False)
plt.title('Количественное соотношение жанров')
plt.axis('equal')
plt.show()
```

Диаграмма "Количественное соотношение жанров" наглядно показывает соотношение жанров контента на платформе Netflix. Для улучшения читабельности диаграммы жанры, доля которых составляет менее 1,5%, были объединены в отдельную группу: *other genres - другие жанры*.
